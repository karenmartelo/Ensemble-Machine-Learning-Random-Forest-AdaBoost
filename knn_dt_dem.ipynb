```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.utils import shuffle
```

```python
# Regression Example
N = 20
Ntrain = 12

# Create sine wave data
X = np.linspace(0, 2*np.pi, N).reshape(N, 1)   # evenly spaced points
Y = np.sin(3*X)                                # target function: sin(3x)
X, Y = shuffle(X, Y)                           # shuffle to randomize train/test split

# Split into training and test sets
Xtrain = X[:Ntrain]
Ytrain = Y[:Ntrain]
```

```python
# Decision Tree Regressor - Low Bias, High Variance
model = DecisionTreeRegressor()                # deep tree (default max_depth=None)
model.fit(Xtrain, Ytrain)

T = 50
Xaxis = np.linspace(0, 2*np.pi, T)             # axis for plotting
Yaxis = np.sin(3*Xaxis)                        # true sine curve

# Plot training points and predictions
plt.scatter(Xtrain, Ytrain, s=50, alpha=0.7, c='blue')   # training data
plt.scatter(Xtrain, model.predict(Xtrain.reshape(Ntrain, 1)), s=50, alpha=0.7, c='green') # predictions on train
plt.title("decision tree - low bias, high variance")

# Plot true curve vs model prediction
plt.plot(Xaxis, Yaxis)
plt.plot(Xaxis, model.predict(Xaxis.reshape(T, 1)))
plt.show()
```

```python
# Decision Tree Regressor - High Bias, Low Variance
model = DecisionTreeRegressor(max_depth=1)     # shallow tree
model.fit(Xtrain, Ytrain)

plt.scatter(Xtrain, Ytrain, s=50, alpha=0.7, c='blue')
plt.scatter(Xtrain, model.predict(Xtrain.reshape(Ntrain, 1)), s=50, alpha=0.7, c='green')
plt.plot(Xaxis, Yaxis)
plt.plot(Xaxis, model.predict(Xaxis.reshape(T, 1)))
plt.title("decision tree - high bias, low variance")
plt.show()
```

```python
# KNN Regressor - Low Bias, High Variance
model = KNeighborsRegressor(n_neighbors=1)     # k=1 nearest neighbor
model.fit(Xtrain, Ytrain)

plt.scatter(Xtrain, Ytrain, s=50, alpha=0.7, c='blue')
```
